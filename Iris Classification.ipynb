### Iris Data import from Sklearn

```python
from sklearn.datasets import load_iris
iris_dataset = load_iris()
```

```python
print("Keys of iris_dataset:")
iris_dataset.keys()
```

```python
print(iris_dataset['DESCR'][:193] + "\n...")
```

```python
print("Target names: {}".format(iris_dataset['target_names']))
print("Feature names: \n{}".format(iris_dataset['feature_names']))
```

```python
print("First five columns of data: \n{}".format(iris_dataset['data'][:5]))
```

Target array contains the species of each of the flowers that were measured, also as a numpy array. 

0 represents *setosa*, 1 means *versicolor*, and 2 means *virginica*.

```python
iris_dataset['target'].shape
```

### Training and Testing Data and Visualizing Data

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(iris_dataset['data'], iris_dataset['target'], random_state=0)
```

```python
print("X_train shape: {}".format(X_train.shape))
print("y_train shape: {}".format(y_train.shape))
print("X_test shape: {}".format(X_test.shape))
print("y_test shape: {}".format(y_test.shape))
```

pair plot of the features in the training set.

To create the plot, we first convert the Numpy Array into a pandas DataFrame. `scatter_matrix`

```python
# create dataframe from data in X_train
iris_dataframe = pd.DataFrame(X_train, columns=iris_dataset.feature_names)

# scatter_matrix from the dataframe
grr = pd.plotting.scatter_matrix(iris_dataframe, c=y_train,figsize=(10,10),
                                 hist_kwds={'bins':20}, cmap=mglearn.cm3,
                                marker='o', alpha=0.8)
```
